```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_excel(
    "../data/raw/online_retail_II.xlsx",
    sheet_name="Year 2010-2011"
)

df.head()
```

1. Check Missing Values

```python
df.isnull().sum()
```

2. Remove Missing Customer Values

```python
df = df.dropna(subset=["Customer ID"])
df.isnull().sum()
```

⬆︎ Dropped The unusable missing values

3. Check Duplicate Records

```python
df.duplicated().sum()
```

4. remove them

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

5. Check for cancelled invoices. they star with a 'C'

```python
df["Invoice"].astype(str).str.startswith("C").sum()
```

6. remove them

```python
df = df[~df["Invoice"].astype(str).str.startswith("C")]
```

7. Remove Invalid Quantities

```python
(df["Quantity"] <= 0).sum()
df = df[df["Quantity"] > 0]
```

8. Remove Invalid Prices

```python
(df["Price"] <= 0).sum()
df = df[df["Price"] > 0]
```

9. INSPECT DATASET AGAIN

```python
df.info()

print("Dataset Shape:", df.shape)
```

10. Save this clean Dataset

```python
df.to_csv(
    "../data/processed/online_retail_clean.csv",
    index=False
)
```
